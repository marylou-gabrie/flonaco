```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import torch
torch.nn.Module.dump_patches = True

import warnings
from torch.serialization import SourceChangeWarning
warnings.filterwarnings("ignore", category=SourceChangeWarning)
```

# Stochastic Allen Cahn experiments $\beta = 20$ and $\beta=10$

## Symmertric models

```python
results = torch.load('../../temp/models/phi4/rnvp_phi4_N100_a0.10_b0.00e+00_beta20.00_rposinit0.10_10blocks.pyT',map_location=torch.device('cpu'))

print('Real-NVP parameters: \n', results['args_model'])

print('Target parameters: \n', results['args_target'])

model = results['model']
print('Final model loaded')

phis = model.sample(10).detach()
phis.shape
for i in range(10):
    plt.plot(np.linspace(0, 1, 100), phis[i, :])
    plt.xlabel('$s$')
    plt.ylabel('$\phi(s)$')
```
